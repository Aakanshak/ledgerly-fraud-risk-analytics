## Calibration and imbalance

Accuracy is misleading: predicting every transaction as legitimate is over 97% accurate yet catches no fraud. ROC-AUC, PR-AUC, precision, recall, calibration, and dollar cost are evaluated on the final two months after training on earlier months.

## So what?

The chosen model is the one that supports lower out-of-time decision cost; model metrics are diagnostics, not the final operating objective.
