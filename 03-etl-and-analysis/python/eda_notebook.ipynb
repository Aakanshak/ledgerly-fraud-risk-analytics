# Ledgerly EDA

Run from the repository root after `python run_pipeline.py`.

```python
import pandas as pd
import plotly.express as px
df=pd.read_parquet('03-etl-and-analysis/outputs/model_features.parquet')
print(df.shape, df.is_fraud.mean())
```

```python
monthly=df.groupby('month').agg(transactions=('transaction_id','size'),fraud_rate=('is_fraud','mean')).reset_index()
px.line(monthly,x='month',y='fraud_rate',markers=True,title='Monthly fraud rate')
```

```python
heat=df.pivot_table(index='merchant_category',columns='ip_country',values='is_fraud',aggfunc='mean')
px.imshow(heat,text_auto='.1%',aspect='auto',color_continuous_scale='RdYlGn_r',title='Fraud rate by category and geography')
```

## So what?

Fraud is rare but concentrated, and risky velocity overlaps legitimate power use. Segment-aware scoring is preferable to single-feature blanket rules.
